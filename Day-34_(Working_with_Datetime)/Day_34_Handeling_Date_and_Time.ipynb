# 1.Now let's perform some operation in date time section

```python
# first convert the column into datetime
date.info()
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
# now extract the year form datetime
date['year'] = date['date'].dt.year
```

```python
# extrat the month
date['month_no'] = date['date'].dt.month
```

```python
# extrat the month name
date["month_name"] = date['date'].dt.month_name()
```

```python
# now extrat the days from the dataset
date['day_no'] = date['date'].dt.day
```

```python
#extract the day name
date['day_name'] = date['date'].dt.day_name()
```

```python
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
# extract the date of week
date['day_no_inweek'] = date['date'].dt.day_of_week
```

# for the weekend

```python
# weended
date['date_is_weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.sample(5)
```

```python
date['week_no'] = date['date'].dt.isocalendar().week
```

```python
#extract quarter
date['quarter_no'] = date['date'].dt.quarter
```

```python
date.head()
```

```python
date['semester_no'] =np.where(date['quarter_no'].isin([1,2]),1,2)
```
