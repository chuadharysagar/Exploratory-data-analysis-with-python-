```python
df.head()
```

```python
data =df.iloc[:,3:5].values
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(12,8))
plt.title("Dendograms")
dend = sch.dendrogram(sch.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

agl = AgglomerativeClustering(n_clusters= 5, affinity= "euclidean" , linkage='ward')
```

```python
labels = agl.fit_predict(data)
```

```python
labels
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data[:,0],data[:,1], c = agl.labels_,cmap="rainbow")
```
