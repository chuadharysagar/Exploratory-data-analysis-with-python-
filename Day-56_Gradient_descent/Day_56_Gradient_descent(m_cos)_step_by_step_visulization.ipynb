```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.predict(X)
```

```python
# slop m
lr.coef_
```

```python
# intercept b
lr.intercept_
```

# NOw in this session let us consider that slop(m) is constant. then we have to find the intercept(b) using gradient descent.

```python
# let's plot the regression line
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

# now lets try the precition with random value

```python
# At initial
b=0
y_pred = ((78.35*X)+b).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,color='green', label ='b={}'.format(b))
plt.plot(X,lr.predict(X) , color='red', label ="OLS")
plt.legend(loc='upper left')
plt.show()
```

```python
slop = -2 * np.sum( y - X.ravel() -b)
slop
```

```python
# here conseder the slop is cosntant m = 78.35
m = 78.35
n =0.1
step_size = n*slop
step_size
```

```python
# calculate the new intercept
b = b - step_size
b
```

```python
#iteration 1
y_pred1 = ((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color= 'red' ,label='OLS')
plt.plot(X,y_pred , color = 'green' , label = 'b = 0')
plt.plot(X,y_pred1,color ='blue',label='iteration 1 b ={}'.format(b))
plt.legend()
plt.show()
```

```python
#iteration 2
slop = -2*np.sum(y-78.35*X.ravel()-b)
slop
```

```python
step_size = n*slop
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35*X)+b).reshape(4)
```

```python
b = -100
m = 78.35

learning_rate = 0.1

epochs = 10

for i in range(epochs):
  loss_slop = -2 *np.sum(y - m*X.ravel() -b)
  step_size = learning_rate * loss_slop
  b = b - step_size
  y_pred  = ((m*X)+b).reshape(4)
  plt.plot(X,y_pred)
plt.scatter(X,y)
plt.show()
```
