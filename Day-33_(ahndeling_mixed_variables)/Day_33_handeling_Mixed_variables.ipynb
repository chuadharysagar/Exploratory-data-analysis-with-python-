## Note: sometime we have to work with the numerical data and categorical data in a single column meaning with the mixed data so in this notebook we will learn to handle the mixed variables

```python
import numpy as np
import pandas as pd
import io
```

```python
from google.colab import  files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['titanic.csv']))
```

```python
df.head()
```

```python
pd.unique(df['number'])
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("Passenger travelling with")
```

```python
# now handle the mixed variable in the number columns
df['number_numerical'] = pd.to_numeric(df['number'] , errors= 'coerce' , downcast= 'integer')
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
#now for the cabin
df['Cabin'].unique()
```

```python
df['Cabin_cat'] = df['Cabin'].str[0]
df['Cabin_num'] = df['Cabin'].str.extract('(\d+)')
```

```python
df.head()
```

```python
# now fo the ticket
# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan , df['ticket_cat'])
```

```python
# extract the second part of ticket as category
df['ticket_num'] = df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'] , errors= 'coerce' , downcast='integer')
```
